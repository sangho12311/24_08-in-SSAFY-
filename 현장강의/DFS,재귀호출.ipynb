# 재귀호출

반복문과 재귀.
재취호출로 n중 반복문을 만들 수 있다. 

### 함수의 특징 
* 함수는 **값만 복사**가 된다. 
* 함수는 호출했던 곳으로 되돌아간다. 


```python
def run(level):
    if level == 3 : # branch==3 
        return
    for i in range(2): # level == 2 
        run(level + 1)

run(0)
```
